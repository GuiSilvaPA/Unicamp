```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
time = np.arange(0.1, 0.25, 0.0001)
vcrit = [4.628 for i in time]
```

```python
t = np.exp((time-0.1)/0.07)/1.5 - 0.5
plt.plot(time, vcrit, label="Vcrit")
plt.plot(time, t, label="Ve")
plt.xlim([0.1, 0.25])
plt.ylim([0, 6])
plt.xlabel("Tempo [s]")
plt.ylabel("Energia Total [pu]")
plt.legend()
plt.grid()
plt.show()
```

```python
plt.plot(time, vcrit, label="Vcrit")
plt.plot(time, t, label="Ve - D = 2")
plt.plot(time, t+0.1, label="Ve - D = 1")
plt.xlim([0.22, 0.245])
plt.ylim([3, 5.5])
plt.xlabel("Tempo [s]")
plt.ylabel("Energia Total [pu]")
plt.legend()
plt.grid()
plt.show()
```

```python
t = np.exp(time/0.2) + 2
plt.plot(time, t, label="Ve")
```
