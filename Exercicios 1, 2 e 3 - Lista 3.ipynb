### Jacobi

```python
x = np.array([[0,0,0]])
print(x)

A = np.array([[10,  3,  1],
              [ 2,-10,  3],
              [ 1,  3, 10]])
print(A)

B = np.array([[14],
              [-5],
              [14]])
print(B)
```

```python
convergencia = False
it = 0
a = [0]
convergiu = ["Não"]
while convergencia == False:
    
    xn = np.zeros(len(x[0]), dtype=float)
    erro = np.zeros(len(x[0]), dtype=float)
    
    for i in range(len(x[0])):
    
        soma = 0
    
        for j in range(len(x[0])):
                
            if j != i:

                soma += A[i,j]*x[it,j]
    
        xn[i] = (1 / A[i,i]) * (B[i] - soma)
    
            
    x = np.vstack([x,xn]) 
    it += 1    
    
    
    
    for i in range(len(x[0])):
    
        erro[i] = np.absolute(x[it,i] - x[it-1,i])
        
    a.append(np.max(erro))
    
    # print(f'{it}:  {x[it,0]}  {x[it,1]}  {x[it,2]} : {a[it]}')
    
    if a[it] < 0.0001:
        convergencia = True
        convergiu.append("Sim")
    else:
        convergiu.append("Não")
```

```python
i = [i for i in range(it+1)]
print(i)

Jacobi_DF = pd.DataFrame(x, columns = ["X1(m)", "X2(m)", "X3(m)"], index = i)

Jacobi_DF["Max"] = a
Jacobi_DF["Convergiu"] = convergiu
```

### Gauss-Seidel

```python
x = np.array([[0,0,0]])
print(x)

A = np.array([[10,  3,  1],
              [ 2,-10,  3],
              [ 1,  3, 10]])
print(A)

B = np.array([[14],
              [-5],
              [14]])
print(B)
```

```python
convergencia = False
it = 0
a = [0]
convergiu = ["Não"]
while convergencia == False:
    
    xn = np.zeros(len(x[0]), dtype=float)
    erro = np.zeros(len(x[0]), dtype=float)
    
    for i in range(len(x[0])):
    
        soma = 0
        soma_f = 0
        
    
        for j in range(i):
            
            soma_f += A[i,j]*xn[j]
    
        for j in range(i+1, len(x[0])):
            
            if j != i:

                soma += A[i,j]*x[it,j]
    
        xn[i] = (1 / A[i,i]) * (B[i] - soma - soma_f)
    
            
    x = np.vstack([x,xn]) 
    it += 1    
    
    for i in range(len(x[0])):
    
        erro[i] = np.absolute(x[it,i] - x[it-1,i])
        
    a.append(np.max(erro))
    
    # print(f'{it}:  {x[it,0]}  {x[it,1]}  {x[it,2]} : {a[it]}')
    
    if a[it] < 0.0001:
        convergencia = True
        convergiu.append("Sim")
    else:
        convergiu.append("Não")
```

```python
i = [i for i in range(it+1)]
print(i)

Gauss_Seidel_DF = pd.DataFrame(x, columns = ["X1(m)", "X2(m)", "X3(m)"], index = i)

Gauss_Seidel_DF["Max"] = a
Gauss_Seidel_DF["Convergiu"] = convergiu
```

1) Ler dados da rede e formar matriz **Y** <br><br>
2) Inicializar contador de iterações *i=0*. Arbitrar valores iniciais das tensões *E* para as barras PQ; arbitrar valores iniciais de tetha para as barras PV.<br><br>
3) Para *k = 1, ...,n*, calcular tensão E<sub>k</sub><sup>i+1</sup>, exceto para a barra slack; e para a barra PV calcular apenas o theta <br><br>
4) Calcular os valores de P<sub>k</sub><sup>esp</sup> para as barras PQ e PV<br><br>
5) Calcular os valores de Q<sub>k</sub><sup>esp</sup> para as barras PQ

6) Realizar os cálculos: <br>

&nbsp;&nbsp;&nbsp;dP<sub>k</sub> = |P<sub>k</sub><sup>calc</sup> - P<sub>k</sub><sup>esp</sup>| para as barras PQ e PV <br>
&nbsp;&nbsp;&nbsp;dQ<sub>k</sub> = |Q<sub>k</sub><sup>calc</sup> - Q<sub>k</sub><sup>esp</sup>| para as barras PQ <br>

7) Obter o valor maximo de mismatches entre todos os valores de *dP<sub>k</sub>* e *dQ<sub>k</sub>*, se ele for menor que uma tolerânicia, o algoritmo se encerra e o estado da rede foi determinado. Caso contrário, deve-se incrementar contador de iterações *i = i + 1* e retornar para passo 3) <br><br>

### a) S2 = 0.3 + 0.3j

```python
Z1 = 0.1 + 0.3j
Z2 = 0.2 + 0.6j

E1 = 1

E2 = [1]
Y = np.zeros((2,2), dtype=np.complex128)

S2 = 0.3 + 0.3j

Zeq = (Z1 * Z2) / (Z1 + Z2)

y12 = 1/Zeq
print(y12)

Y[0,0] = y12
Y[0,1] = -y12
Y[1,0] = -y12
Y[1,1] = y12
```

```python
convergencia = False
i = 0
tolerancia = 0.0001

S2_final = np.conj(Y[1,0]*E1 + Y[1,1]*E2[i]) * E2[i]

print(f'{i} |  V2 = {np.absolute(E2[i]):.5f}  |  O = {np.angle(E2[i],deg=True):.5f}°  |  P2 = {np.real(S2_final):.5f}  |  Q2 = {np.imag(S2_final):.5f}')

while convergencia == False:
    

    E2_new = (1/Y[1,1]) * (np.conj(S2)/np.conj(E2[i]) - Y[1,0]*E1)
    # print(E2_new)
    
    E2.append(E2_new)
    
    i += 1
    
    S2_final = np.conj(Y[1,0]*E1 + Y[1,1]*E2[i]) * E2[i]
    
    print(f'{i} |  V2 = {np.absolute(E2[i]):.5f}  |  O = {np.angle(E2[i],deg=True):.5f}°  |  P2 = {np.real(S2_final):.5f}  |  Q2 = {np.imag(S2_final):.5f}')
    
    if tolerancia > np.absolute(E2[i] - E2[i-1]):
        convergencia = True
        
```
